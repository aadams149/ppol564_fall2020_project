```python
#Read in the finalized dataset
df = pd.read_csv('aja149_cleandata.csv')
```

```python
#First, check the dtypes
df.dtypes
```

```python
#First, the measures of Rural/Urbanness
d = df[['Rural_urban_continuum_code_2013','Urban_influence_code_2013','Metro_2013']].melt()

(
    ggplot(d,aes(x="value")) +
    geom_histogram(bins=25) +
    facet_wrap("variable",scales='free') +
    theme(figure_size=(10,3),
          subplots_adjust={'wspace':0.25})
)
```

```python
#Next, the unemployment rates across different years
d = df.filter(regex='Unemployment').melt()

(
    ggplot(d,aes(x="value")) +
    geom_histogram(bins=25) +
    facet_wrap("variable",scales='free') +
    theme(figure_size=(20,12),
          subplots_adjust={'wspace':0.50})
)
```

```python
#Next, the data on proportions of the 2-party vote share
d = df.filter(regex='votes').melt()

(
    ggplot(d,aes(x="value")) +
    geom_histogram(bins=25) +
    facet_wrap("variable",scales='free') +
    theme(figure_size=(20,12),
          subplots_adjust={'wspace':0.50})
)
```

```python
#Next, the data on proportions of the 2-party vote share
d = df.filter(regex='2P').melt()

(
    ggplot(d,aes(x="value")) +
    geom_histogram(bins=25) +
    facet_wrap("variable",scales='free') +
    theme(figure_size=(20,12),
          subplots_adjust={'wspace':0.50})
)
```

```python
#Next, the data on race and gender
d = df.filter(regex='MALE').melt()

(
    ggplot(d,aes(x="value")) +
    geom_histogram(bins=25) +
    facet_wrap("variable",scales='free') +
    theme(figure_size=(20,12),
          subplots_adjust={'wspace':0.50})
)
```

```python
#Finally, the dependent variable: which party's presidential candidate won a county in 2016
d = df.filter(regex='win').melt()

(
    ggplot(d,aes(x="value")) +
    geom_histogram(bins=25) +
    facet_wrap("variable",scales='free') +
    theme(figure_size=(20,12),
          subplots_adjust={'wspace':0.50})
)
```
